The function should return the sub-array and the corresponding sum. Which is the complexity of your solution? Can you do a solution with linear complexity? Write a unit test for the function.

```python
def contiguous_subarray(array, size):
    '''
        The function receives an array of numbers and a size of the subarray and returns the largest sum
        of all the subarrays in the initial arrays
        
        Params.
        -------
        array = array of integers
        size = integer
        
        Returns.
        --------
        The largest sum and the subarray that gives this sum
    '''
    if size == 0 or len(array)<size:
        if size == 0:
            return print('Please introduce de correct params, array len will be greater than subarray size')
        if len(array)<size:
            return print(f'The size of subarray "{size}" is greater than size of the initial array "{len(array)}"')
    else:
        solution_array=[]
        for i in range((len(array)+1-size)+1):
            temp = array[i:i+size]
            if i>0 and sum(solution_array)<sum(temp):
                solution_array=temp
    
    return print(f'The largest sum of the array given is {sum(solution_array)} and the subarray is {solution_array}')
                
```
